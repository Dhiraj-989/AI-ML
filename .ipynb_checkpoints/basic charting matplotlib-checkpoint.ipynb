```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
x = [2,4,5]
y = [4,6,7]
x2 = [3,8,9]
y2 = [ 22,4,55]

plt.title("DATA")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.grid(True,color = 'k')

plt.plot(x, label = "first", linewidth = 3)
plt.plot(x2, label = "second", linewidth = 3)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
style.use('ggplot')
plt.bar([1,3,4,6],[2,7,6,8],label="Example one",linewidth = 3)
plt.bar([3,4,5,7],[3,4,7,8],label="Example two",linewidth = 3,color = 'g')
plt.legend()
plt.xlabel("Bar Number")
plt.ylabel("Bar Height")
plt.show()
```

```python
days = [1,2,3,4,5]
waking = [2,4,6,8,0]
sleeping = [3,4,7,9,6]
walking = [2,3,7,9,4]


plt.plot([],[], color = 'm' ,label = 'waking' , linewidth = 5)
plt.plot([],[], color = 'k' , label = 'sleeping' ,linewidth = 5)
plt.plot([],[], color = 'c' ,label = 'walking' , linewidth = 5)

plt.stackplot(days,waking,sleeping,walking,colors = ['m','k','c'])
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data to plot
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [35, 30, 20, 15]  # Percentage distribution
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Creating the pie chart
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode, shadow=True)
plt.title("Programming Language Usage")
plt.show()
```
