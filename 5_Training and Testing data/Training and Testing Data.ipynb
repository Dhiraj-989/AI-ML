# Using train test split method from sklearn
Import train_test_split from sklearn

call train_test_split method

and provide the test size 

in this case test size is 20% of the dataset and training size is the rest which is 80%

we get four parameters i.e. x_train, x_test, y_train, y_test

specify randoom state so that it does not take random data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state = 10)
```
