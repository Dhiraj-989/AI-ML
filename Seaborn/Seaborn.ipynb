```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a = sns.load_dataset("flights")
sns.relplot(x="passengers", y="month", data=a)
plt.show()
```

```python
a = sns.load_dataset("flights")
sns.relplot(x="month", y="passengers", data=a, kind = "line")
plt.show()
```

```python
b = sns.load_dataset("tips")
sns.relplot(x = "time" , y = "tip", data = b, kind = "line")
```

```python
sns.catplot(x = "day", y = "total_bill", data = b)
```

```python
sns.catplot(x="day", y="total_bill", kind="violin", palette="pastel", hue="sex", split=True, data=b)
```
